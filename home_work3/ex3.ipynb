```python
import numpy as np
from math import factorial
import itertools as it
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

#### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
X = np.array((100,80, 75,77,89,33,45,25,65,17,30,24,57,55,70,75,65,84,90,150))
X
```

```python
mean = np.sum(X)/len(X) 
print(mean)
print(X.mean())
```

```python
sko = np.sqrt(np.sum((X - mean)**2)/len(X))
print(sko)
print(X.std())
```

```python
# смещенная оценка дисперсии
S_2 = np.sum((X - mean)**2)/len(X)
print(S_2)
print(X.var())
```

```python
# несмещенная оценка дисперсии
S_2 = np.sum((X - mean)**2)/(len(X)-1)
print(S_2)
print(X.var(ddof=1))
```

#### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
# Возможны след ситуации:  
# P1 - из первого ящика вытаскиваем 2 белых мяча 0 черных, из второго - 1б 3ч
# Р2 - из первого - 1б 1ч, из второго - 2б 2ч
# Р3 - из первого - 0б 2ч, из второго - 3б 1ч
P1 = combinations(5,2)/combinations(8,2) * combinations(5,1)*combinations(7,3)/combinations(12,4)
print(f'P1 = {P1}')
P2 = combinations(5,1)*combinations(3,1)/combinations(8,2) * combinations(5,2)*combinations(7,2) \
    /combinations(12,4)
print(f'P2 = {P2}')
P3 = combinations(3,2)/combinations(8,2) * combinations(5,3)/combinations(12,4)
print(f'P3 = {P3}')
print(f'Вероятность того, что 3 мяча белые: {(P1+P2+P3):.3f}')
```

#### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# расчет производить будем с помощью формулы Байеса
# полная вероятность P_A
P_A = (1/3*0.9 + 1/3*0.8 + 1/3*0.6)
# выстрел произведен первым спортсменом
P_1 = (1/3*0.9)/P_A
print(f'Вероятность того, что выстрел произведен первым спортсменом: {P_1*100:.0f}%')
print(f'({P_1:.7f})')

P_2 = (1/3*0.8)/P_A
print(f'Вероятность того, что выстрел произведен первым спортсменом: {P_2*100:.0f}%')
print(f'({P_2:.7f})')

P_3 = (1/3*0.6)/P_A
print(f'Вероятность того, что выстрел произведен первым спортсменом: {P_3*100:.0f}%')
print(f'({P_3:.7f})')
```

#### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# расчет производить будем с помощью формулы Байеса
# вероятность того, что студент из группы А - 1/4, из В - 1/4, из С - 2/4 (или 1/2) 
P1 = 1/4*0.8
P2 = 1/4*0.7
P3 = 2/4*0.9
# полная вероятность P_A
P_A = P1 + P2 + P3

P_1 = P1/P_A
print(f'Вероятность того, что студент учится на факультете A: {P_1*100:.0f}%')
print(f'({P_1:.7f})')

P_2 = P2/P_A
print(f'Вероятность того, что  студент учится на факультете В: {P_2*100:.0f}%')
print(f'({P_2:.7f})')

P_3 = P3/P_A
print(f'Вероятность того, что  студент учится на факультете С: {P_3*100:.0f}%')
print(f'({P_3:.7f})')
```

#### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25

P = p1*p2*p3
print(f'a: Вероятность того, что выйдут из строя все детали: {P*100:.0f}%')
print(f'({P:.7f})')

P = p1*p2*(1-p3) + p1*p3*(1-p2) + p2*p3*(1-p1)
print(f'б: Вероятность того, что выйдут из строя только две детали: {P*100:.0f}%')
print(f'({P:.7f})')

P = 1 - (1-p1)*(1-p2)*(1-p3)
print(f'в: Вероятность того, что выйдет из строя хотя бы одна деталь: {P*100:.0f}%')
print(f'({P:.7f})')

P = p1*p2*(1-p3) + p1*p3*(1-p2) + p2*p3*(1-p1) + p1*(1-p2)*(1-p3) + p2*(1-p1)*(1-p3) + p3*(1-p2)*(1-p1)
print(f'г: Вероятность того, что выйдет из строя от одной до двух деталей: {P*100:.0f}%')
print(f'({P:.7f})')
```
